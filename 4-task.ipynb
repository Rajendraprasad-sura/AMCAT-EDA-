## uivariate analysis

```python
# PDFs and Histograms for Numerical Columns
for col in df_n:
    plt.figure(figsize=(5, 3))
    sns.histplot(df1[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Boxplots for Numerical Columns
for col in df_n:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df1[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
for col in df_o:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=df1[col])
    plt.title(f'Countplot of {col}')
    plt.show()
```

```python
# Create count plots for the top 10 values in each categorical column
for col in df_o:
    # Get the top 10 values and their counts
    top_values = df1[col].value_counts().head(10)
    
    # Create a count plot for the top 10 values
    plt.figure(figsize=(4, 3))
    sns.countplot(x=col, data=df1, order=top_values.index)
    plt.title(f'Count Plot for Top 10 Values in {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Identify numerical columns (float and int)
numerical_columns = df1.select_dtypes(include=['float64', 'int64']).columns

# Find outliers using IQR for each numerical column
outliers_dict = {}
for col in numerical_columns:
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers_indices = (df1[col] < lower_bound) | (df1[col] > upper_bound)
    outliers_dict[col] = df1[col][outliers_indices]

# Display the outliers
for col, outliers in outliers_dict.items():
    print(f'Outliers in {col}:\n{outliers}\n')
```
